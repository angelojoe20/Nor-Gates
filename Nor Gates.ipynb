```python
#Program Title: Basic Logic Gates - Nor Gates

import matplotlib.pyplot as plt

truth_table = range(4)
x = [0, 0, 1, 1]
y = [0, 1, 0, 1]
z = [1, 0, 0, 0]

plt.subplot(2, 1, 1)
plt.step(truth_table, x, where='post', label='x')
plt.step(truth_table, y, where='post', label='y')
plt.xlabel('Logic Gates - Nor Gates Waveform')
plt.ylabel('A')
plt.yticks([0, 1])
plt.legend()

plt.subplot(2, 1, 2)
plt.step(truth_table, z, where='post', label='z')
plt.xlabel('Logic Gates - Nor Gates Waveform')
plt.ylabel('B')
plt.yticks([0, 1])
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Program Title: Full Adder using Nor Gates

import matplotlib.pyplot as plt

truth_table = range(8)
a = [0, 0, 0, 0, 1, 1, 1, 1]
b = [0, 0, 1, 1, 0, 0, 1, 1]
cin = [0, 1, 0, 1, 0, 1, 0, 1]
s = [0, 1, 1, 0, 1, 1, 0, 1]
cout = [0, 0, 0, 1, 0, 1, 1, 1]

plt.subplot(4, 1, 1)
plt.step(truth_table, a, where='post', label='a')
plt.step(truth_table, b, where='post', label='b')
plt.title('Full Adder using Nor Gates Waveform')
plt.ylabel('A & B')
plt.yticks([0, 1])
plt.legend()

plt.subplot(4, 1, 2)
plt.step(truth_table, cin, where='post', label='carry in',color='r')
plt.ylabel('Carry In')
plt.yticks([0, 1])
plt.legend()

plt.subplot(4, 1, 3)
plt.step(truth_table, s, where='post', label='sum', color='y')
plt.ylabel('Sum')
plt.yticks([0, 1])
plt.legend()

plt.subplot(4, 1, 4)
plt.step(truth_table, cout, where='post', label='carry out', color='g')
plt.ylabel('Carry Out')
plt.yticks([0, 1])
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Program Title: Half Adder using Nor Gates

import matplotlib.pyplot as plt

truth_table = range(4)
a = [0, 0, 1, 1]
b = [0, 1, 0, 1]
sum = [0, 1, 1, 0]
carry = [0, 0, 0, 1]

plt.subplot(3, 1, 1)
plt.step(truth_table, a, where='post', label='a')
plt.step(truth_table, b, where='post', label='b')
plt.title('Half Adder using Nor Gates WaveForm')
plt.ylabel('Input')
plt.yticks([0, 1])
plt.legend()


plt.subplot(3, 1, 2)
plt.step(truth_table, sum, where='post', label='Sum', color='g')
plt.ylabel('Output')
plt.yticks([0, 1])
plt.legend()


plt.subplot(3, 1, 3)
plt.step(truth_table, carry, where='post', label='Carry', color='r')
plt.ylabel('Output')
plt.yticks([0, 1])
plt.legend()
```
